```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mall_data = pd.read_csv("Mall_Customers.csv")
```

```python
mall_data.head()
```

```python
mall_data.shape
```

```python
#missing values
mall_data.info()
```

```python
X = mall_data.iloc[:,[3,4]].values
X
```

```python
from sklearn.cluster import KMeans
wcss=[]
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("Number of clusters")
plt.ylabel("WCSS value")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++',random_state = 0)
```

```python
y = kmeans.fit_predict(X)
```

```python
y
```

```python
plt.figure(figsize = (8,8))
#ax = fig.add_subplot(111, projection='3d')
plt.scatter(X[y == 0,0],X[y == 0,1], s = 40 , color = 'red', label = "cluster 1")
plt.scatter(X[y == 1,0],X[y == 1,1], s = 40 , color = 'blue', label = "cluster 2")
plt.scatter(X[y == 2,0],X[y == 2,1], s = 40 , color = 'green', label = "cluster 3")
plt.scatter(X[y == 3,0],X[y == 3,1], s = 40 , color = 'yellow', label = "cluster 4")
plt.scatter(X[y == 4,0],X[y == 4,1], s = 40 , color = 'purple', label = "cluster 5")
plt.xlabel('Anual Income-->')
plt.ylabel('Spending Score-->')
#ax.legend()
plt.show()
```

```python
# elbow graph
sns.set()
plt.plot(range(1,11),wcss)
plt.xlabel("Number of clusters")
plt.ylabel("WCSS value")
plt.show()
```
